```
import pandas as pd


df = pd.read_csv("Datase.csv")
df.head()
```

```
exceptional = df[df["Score"] > 90]
print(exceptional)
```

```
sales = df[df["Department"] == "Sales"]
print(sales)
```

```
q1 = df[df["Quarter"] == "Q1"]
print(q1)
```

```
sorted_score = df.sort_values(by = "Score", ascending=False)
print("Sorted by Score:\n", sorted_score)
```

```
sorted_dept = df.sort_values(by=["Department", "Score"], ascending=[True, False])
print("Department Ranking:\n", sorted_dept)
```

```
avg_dept = df.groupby("Department")["Score"].mean()
print("Average Score per Department:\n", avg_dept)
```

```
max_quarter = df.groupby("Quarter")["Score"].max()
print("Max Score per Quarter:\n", max_quarter)
```
