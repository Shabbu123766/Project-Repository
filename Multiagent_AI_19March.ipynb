```
# Example 1: Sequential Multi-Agent System (Pipeline)
# Scenario

# “A travel company has different employees (agents):

# Planner → decides steps

# Flight Agent → finds flights

# Weather Agent → checks weather

# Decision Agent → gives final answer”

# ================================
# AGENT 1: PLANNER
# ================================
def planner_agent(user_query):
    print("\n[Planner Agent] Creating plan...")
    return ["flight", "weather", "decision"]


# ================================
# AGENT 2: FLIGHT AGENT
# ================================
def flight_agent():
    print("\n[Flight Agent] Fetching flights...")
    return [
        {"airline": "IndiGo", "price": 4500},
        {"airline": "Air India", "price": 5200}
    ]


# ================================
# AGENT 3: WEATHER AGENT
# ================================
def weather_agent():
    print("\n[Weather Agent] Checking weather...")
    return {"condition": "Clear", "temp": 28}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(flights, weather):
    print("\n[Decision Agent] Making decision...")

    cheapest = min(flights, key=lambda x: x["price"])

    if weather["condition"] == "Rain":
        return "Avoid travel due to bad weather"

    return f"Book {cheapest['airline']} at ₹{cheapest['price']}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def travel_multi_agent(user_query):
    print("User Query:", user_query)

    plan = planner_agent(user_query)

    flights = None
    weather = None

    for step in plan:
        if step == "flight":
            flights = flight_agent()

        elif step == "weather":
            weather = weather_agent()

        elif step == "decision":
            result = decision_agent(flights, weather)

    return result


# RUN
response = travel_multi_agent("Plan my trip Delhi to Mumbai")
print("\nFinal Answer:", response)
```

```
# SCenario
# “A hospital uses different employees (agents) to handle patient care in sequence.”

# ================================
# AGENT 1: Intake Agent (Planner)
# - Collects patient symptoms and history
# - Decides which steps are needed (tests, consultations, etc.)

# ================================
# AGENT 2: Diagnostic Agent
# - Orders lab tests or scans
# - Interprets results and identifies possible conditions

# ================================
# AGENT 3: Treatment Agent
# - Suggests treatment options (medication, therapy, surgery)
# - Considers patient preferences and medical guidelines

# ================================
# AGENT 4: Decision Agent
# - Reviews all inputs (history, diagnostics, treatment options)
# - Provides the final recommendation to the patient

# ================================
# AGENT 1: INTAKE AGENT (PLANNER)
# ================================
def intake_agent(patient_query):
    print("\n[Intake Agent] Collecting patient details...")
    return {
        "symptoms": patient_query,
        "plan": ["diagnosis", "treatment", "decision"]
    }


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent(symptoms):
    print("\n[Diagnostic Agent] Analyzing symptoms...")

    if "fever" in symptoms.lower() and "cough" in symptoms.lower():
        return {"diagnosis": "Possible Infection"}
    else:
        return {"diagnosis": "General Checkup Required"}


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent(diagnosis):
    print("\n[Treatment Agent] Suggesting treatment...")

    if "infection" in diagnosis.lower():
        return {"treatment": "Prescribe antibiotics"}
    else:
        return {"treatment": "Rest and hydration"}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(diagnosis, treatment):
    print("\n[Decision Agent] Final recommendation...")

    return f"Diagnosis: {diagnosis} | Treatment: {treatment}"


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def hospital_multi_agent(patient_query):
    print("Patient Query:", patient_query)

    intake_data = intake_agent(patient_query)

    diagnosis = None
    treatment = None

    for step in intake_data["plan"]:

        if step == "diagnosis":
            diag_result = diagnostic_agent(intake_data["symptoms"])
            diagnosis = diag_result["diagnosis"]

        elif step == "treatment":
            treat_result = treatment_agent(diagnosis)
            treatment = treat_result["treatment"]

        elif step == "decision":
            result = decision_agent(diagnosis, treatment)

    return result


# RUN
response = hospital_multi_agent("Patient has fever and cough")
print("\nFinal Answer:", response)
```

```
import requests

# ================================
# API CONFIG
# ================================
API_TOKEN = "hf_token"
MODEL_ID = "meta-llama/Meta-Llama-3-8B-Instruct"

API_URL = "https://router.huggingface.co/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_TOKEN}",
    "Content-Type": "application/json"
}


# ================================
# LLM QUERY FUNCTION
# ================================
def query_llm(prompt: str):
    payload = {
        "model": MODEL_ID,
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "max_tokens": 150
    }

    response = requests.post(API_URL, headers=headers, json=payload)
    data = response.json()

    if "choices" in data:
        return data["choices"][0]["message"]["content"]
    else:
        return str(data)


# ================================
# AGENT 1: INTAKE AGENT (PLANNER)
# ================================
def intake_agent(patient_query):
    print("\n[Intake Agent] Understanding patient...")

    thought = query_llm(f"Analyze patient symptoms: {patient_query}")
    print("LLM Thought:", thought)

    return {
        "symptoms": patient_query,
        "plan": ["diagnosis", "treatment", "decision"]
    }


# ================================
# AGENT 2: DIAGNOSTIC AGENT
# ================================
def diagnostic_agent(symptoms):
    print("\n[Diagnostic Agent] Diagnosing...")

    diagnosis = query_llm(f"What could be the diagnosis for: {symptoms}?")
    print("LLM Diagnosis:", diagnosis)

    return {"diagnosis": diagnosis}


# ================================
# AGENT 3: TREATMENT AGENT
# ================================
def treatment_agent(diagnosis):
    print("\n[Treatment Agent] Suggesting treatment...")

    treatment = query_llm(f"Suggest treatment for: {diagnosis}")
    print("LLM Treatment:", treatment)

    return {"treatment": treatment}


# ================================
# AGENT 4: DECISION AGENT
# ================================
def decision_agent(diagnosis, treatment):
    print("\n[Decision Agent] Final decision...")

    final = query_llm(
        f"Based on diagnosis: {diagnosis} and treatment: {treatment}, give final recommendation."
    )

    return final


# ================================
# MAIN MULTI-AGENT SYSTEM
# ================================
def hospital_multi_agent(patient_query):
    print("Patient Query:", patient_query)

    intake_data = intake_agent(patient_query)

    diagnosis = None
    treatment = None

    for step in intake_data["plan"]:

        if step == "diagnosis":
            diag_result = diagnostic_agent(intake_data["symptoms"])
            diagnosis = diag_result["diagnosis"]

        elif step == "treatment":
            treat_result = treatment_agent(diagnosis)
            treatment = treat_result["treatment"]

        elif step == "decision":
            result = decision_agent(diagnosis, treatment)

    return result


# ================================
# RUN
# ================================
response = hospital_multi_agent("Patient has fever and cough")
print("\nFinal Answer:\n", response)
```

```
# Real-World Pipeline: AI Research Team
#  Goal:

# “Write a comprehensive market report on EV industry trends in India for 2025.”

#  Pipeline Components
#  Orchestrator

# Splits & assigns tasks

#  Search Agent

# Finds EV data online

#  Analyst Agent

# Interprets & summarises

#  Writer Agent

# Drafts the report

#  QA Agent

# Checks facts & tone

# Shared Memory / Message Bus

# All agents read & write findings here in real time

#  Orchestration Code (from image)
# # Orchestrating the Research Pipeline

# agents = [search, analyst, writer, qa]
# memory = SharedMemory()

# for agent in agents:
#     result = agent.run(goal, memory)
#     memory.update(agent.name, result)

# report = memory.get("writer")
#  Why Multi-Agent Wins Here

# Parallel execution → 4x faster

# Specialists = higher quality output

# QA agent catches errors automatically

# Easy to swap or add new agents

# Scales to 10x more tasks with no change



from groq import Groq

# ================================
#  API SETUP
# ================================
client = Groq(api_key="api_key")


# ================================
#  LLM FUNCTION
# ================================
def query_llm(prompt):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.5
    )
    return response.choices[0].message.content


# ================================
#  SHARED MEMORY
# ================================
class SharedMemory:
    def __init__(self):
        self.data = {}

    def update(self, agent_name, result):
        print(f"\n[Memory] {agent_name} updated")
        self.data[agent_name] = result

    def get(self, agent_name):
        return self.data.get(agent_name, "")

    def get_all(self):
        return self.data


# ================================
#  SEARCH AGENT
# ================================
class SearchAgent:
    name = "search"

    def run(self, goal, memory):
        print("\n[Search Agent]")
        return query_llm(f"Find key data and trends about: {goal}")


# ================================
#  ANALYST AGENT
# ================================
class AnalystAgent:
    name = "analyst"

    def run(self, goal, memory):
        print("\n[Analyst Agent]")
        search_data = memory.get("search")

        return query_llm(f"""
        Analyze the following data and summarize insights:

        {search_data}
        """)


# ================================
#  WRITER AGENT
# ================================
class WriterAgent:
    name = "writer"

    def run(self, goal, memory):
        print("\n[Writer Agent]")
        analysis = memory.get("analyst")

        return query_llm(f"""
        Write a detailed market report based on this analysis:

        {analysis}
        """)


# ================================
#  QA AGENT
# ================================
class QAAgent:
    name = "qa"

    def run(self, goal, memory):
        print("\n[QA Agent]")
        report = memory.get("writer")

        return query_llm(f"""
        Review the following report for correctness, clarity, and tone.
        Improve if needed and give final version:

        {report}
        """)


# ================================
#  ORCHESTRATOR
# ================================
class Orchestrator:
    def __init__(self):
        self.agents = [
            SearchAgent(),
            AnalystAgent(),
            WriterAgent(),
            QAAgent()
        ]
        self.memory = SharedMemory()

    def run(self, goal):
        print(" Goal:", goal)

        for agent in self.agents:
            result = agent.run(goal, self.memory)
            self.memory.update(agent.name, result)

        final_report = self.memory.get("qa")
        return final_report


# ================================
#  RUN
# ================================
goal = "EV industry trends in India for 2025"

orchestrator = Orchestrator()
final_output = orchestrator.run(goal)

print("\n FINAL REPORT:\n", final_output)

```
