```
import pandas as pd


df = pd.read_csv('Dataset.csv')
```

```
# Preview data
print("Dataset Preview")
print(df.head())
```

```
# Employees with Score > 90
exceptional = df[df['Score'] > 90]
print("\nExceptional Performers (Score > 90)")
print(exceptional)
```

```
# Sales department only
sales_dept = df[df['Department'] == 'Sales']
print("\nSales Department Records")
print(sales_dept)
```

```
# Sort by Score descending
sorted_score = df.sort_values(by='Score', ascending=False)
print("\nEmployees Ranked by Score")
print(sorted_score)

# Sort by Department then Score
dept_ranking = df.sort_values(by=['Department','Score'], ascending=[True, False])
print("\nDepartment-wise Ranking")
print(dept_ranking)
```

```
# Average score per department
avg_dept = df.groupby('Department', as_index=False)['Score'].mean()
print("\nAverage Score per Department")
print(avg_dept)

# Maximum score per quarter
max_quarter = df.groupby('Quarter', as_index=False)['Score'].max()
print("\nMaximum Score per Quarter")
print(max_quarter)
```
