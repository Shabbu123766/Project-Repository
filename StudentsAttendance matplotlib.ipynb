```
#Description
 #📘 Scenario Question
 # A college is tracking student attendance across the first four months of the semester:
 # - January: 85 students present on average # - February: 90 students # - March: 95 students
 # - April: 88 students # The academic coordinator wants to visualize the attendance trend to see whether students are attending classes more regularly or
 # if there are fluctuations. # You are asked to:
  # - Plot a line chart showing the monthly attendance trend.
  # - Label the axes and add a title to make the chart clear.
  # - Interpret the chart: # - Which month had the highest attendance?
  # - Did attendance steadily increase, or were there ups and downs?
  # - What might explain the dip in April?

import matplotlib.pyplot as plt
```

```
months = ["Jan", "Feb", "Mar", "Apr"]
attendance = [85, 90, 95, 88]

plt.plot(months, attendance)
plt.title("Attendance chart")
plt.xlabel("Months")
plt.ylabel("Attendance")
plt.show()
```
