```
sales_day1 = np.array([120, 85, 60])
sales_day2 = np.array([150, 90, 75])
total_sales = sales_day1 + sales_day2
print(total_sales)
```

```
growth = ((sales_day2 - sales_day1)/sales_day1)*100
print(growth)
```

```
products = ["Product A", "Product B", "Product C"]
highest_index = np.argmax(growth)
print("highest_growth:", products[highest_index])
```

```
yields = np.array([[2.5, 3.0, 1.0], [2.8, 3.2, 2.0], [3.1, 2.9, 2.2], [2.6, 3.1, 1.9]])
yields[1] = yields[1] + np.array([0.2, 0.3, 0.1])
print(yields)
```

```
yields[:, 2] = yields[:, 2] * 1.1
print(yields)
```

```
avg = np.mean(yields, axis=0)
crops = ["CROP A", "CROP B", "CROP C"]
print("highest_avg:", crops[np.argmax(avg)])
```

```
# Scenario: Farm Crop Yield Analysis
import numpy as np

yields = np.array([[100, 200, 300, 400], [500, 600, 700, 800], [900, 500, 400, 300]])

yields_increased = yields * 0.1
print("yields_increased by 10% : \n", yields_increased)
```

```
yields_increased = yields_increased[0]
print("Only wheat's yield:\n", yields_increased)
```

```
avg_per_crop = np.mean(yields_increased, axis = 1)
print("Average yield per crop across all seasons:\n", avg_per_crop)
```

```
yields_increased = yields_increased[yields_increased > 30]
print("yields greater than 300:\n", yields_increased)
```

```
#scenario : Factory Production Analaysis
import numpy as np

production = np.array([[3425, 3426, 7654, 5647], [2314, 2876, 5634, 9087], [3425, 3454, 7865, 4536]])

increased_production = production * 0.05
print("Production increased by 5%:\n", increased_production)
```

```
increased_production = increased_production[0]
print("Gadget's production:\n", increased_production)
```

```
avg_per_product = np.mean(increased_production, axis = 1)
print("Average production per product:\n", avg_per_product)
```

```
increased_production = increased_production[increased_production > 50]
print("Values greater than 50:\n", increased_production)
```
