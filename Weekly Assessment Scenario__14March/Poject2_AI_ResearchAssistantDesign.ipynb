```
#Task 2 – Prompt Engineering Experiments (8 Marks)
# Design and test three different prompting strategies.

# 1️⃣ Zero-Shot Prompt
# Ask the model to summarize the article without providing examples.

# Example task:

# Summarize the following article in 5 bullet points.

# 2️⃣ Few-Shot Prompt
# Provide 2 example summaries before asking the model to summarize a new article.

# 3️⃣ Chain-of-Thought Prompt
# Ask the model to reason step-by-step before generating the summary.

# Example structure:

# Analyze the article step by step:

# 1. Identify the main topic

# 2. Extract key ideas

# 3. Generate a concise summary

# Deliverable:

# All three prompts


# Model outputs


# Comparison of responses


from transformers import pipeline

# Load LLM
generator = pipeline("text-generation", model="gpt2")

# Article to summarize
article = """
Artificial Intelligence is transforming industries such as healthcare,
education, finance, and transportation. Machine learning algorithms allow
computers to analyze large datasets and identify patterns. In healthcare,
AI helps doctors detect diseases earlier through medical imaging analysis.
In education, AI systems provide personalized learning experiences.
However, the rapid growth of AI also raises concerns related to privacy,
bias in algorithms, and job displacement.
"""


# 1: Zero-Shot Prompt


zero_shot_prompt = f"""
Summarize the following article in 5 bullet points:

{article}
"""

zero_output = generator(zero_shot_prompt, max_length=150)

print("------ ZERO SHOT PROMPT ------\n")
print(zero_output[0]["generated_text"])



# 2: Few-Shot Prompt


few_shot_prompt = f"""
Example 1
Article: Solar energy is renewable and helps reduce pollution.
Summary:
• Solar energy is renewable
• Helps reduce carbon emissions
• Supports sustainable energy

Example 2
Article: Online education allows students to learn from anywhere.
Summary:
• Flexible learning
• Accessible globally
• Growth of digital education

Now summarize the following article in 5 bullet points:

{article}
"""

few_output = generator(few_shot_prompt, max_length=180)

print("\n\n------ FEW SHOT PROMPT ------\n")
print(few_output[0]["generated_text"])



# 3: Chain-of-Thought Prompt


cot_prompt = f"""
Analyze the article step by step.

1. Identify the main topic
2. Extract key ideas
3. Generate a short summary

Article:
{article}
"""

cot_output = generator(cot_prompt, max_length=180)

print("\n\n------ CHAIN OF THOUGHT PROMPT ------\n")
print(cot_output[0]["generated_text"])
```

```
# Task 3 – Prompt Optimization (7 Marks)
# Improve the prompt so that the AI generates better executive summaries.

# Your optimized prompt must produce:

# 3 key insights


# 1 actionable takeaway


# Professional tone


# Test the optimized prompt on two different articles.

# Deliverable:

# Final optimized prompt


# Two example outputs


from transformers import pipeline

# Load LLM model
generator = pipeline("text-generation", model="gpt2")

# -------- Optimized Prompt Template --------

def generate_summary(article):

    prompt = f"""
You are an AI research assistant.

Read the following article and generate an executive summary.

Output format:

Key Insights:
1.
2.
3.

Actionable Takeaway:

Tone: Professional

Article:
{article}
"""

    result = generator(prompt, max_length=200)
    return result[0]["generated_text"]


# -------- Article 1 --------

article1 = """
Artificial Intelligence helps organizations automate repetitive tasks
and analyze large datasets efficiently. Businesses use AI to improve
decision-making and customer experiences. However, organizations must
ensure responsible AI deployment to prevent bias and protect user data.
"""

print("----- ARTICLE 1 OUTPUT -----\n")
print(generate_summary(article1))


# -------- Article 2 --------

article2 = """
Climate change is increasing due to greenhouse gas emissions.
Renewable energy sources such as solar and wind power can help
reduce carbon footprints and promote sustainable development.
Governments and industries are investing in clean energy technologies.
"""

print("\n\n----- ARTICLE 2 OUTPUT -----\n")
print(generate_summary(article2))
```

```
# Task 4 – Tokenization Experiment (5 Marks)
# Use a tokenizer from HuggingFace or any LLM library.

# Steps:

# Tokenize a paragraph of text.


# Count the total number of tokens.


# Display the tokenized output.


# Example tools:

# tiktoken


# transformers tokenizer


# Deliverable:

# Python code


# Token count


# Example tokenized output


from transformers import AutoTokenizer

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Sample paragraph
text = """
Artificial Intelligence is transforming healthcare, education,
and finance by enabling machines to analyze data and make decisions.
"""

# Tokenize text
tokens = tokenizer.tokenize(text)

# Convert tokens to IDs
token_ids = tokenizer.encode(text)

# Print results
print("Original Text:\n", text)

print("\nTokenized Output:")
print(tokens)

print("\nToken IDs:")
print(token_ids)

print("\nTotal Number of Tokens:", len(tokens))
```

```
# Task 5 – Build a Mini AI Tool (5 Marks)
# Create a simple AI assistant that takes a long article and produces:

# A short summary


# Key insights


# One actionable recommendation


# The program should:

# Accept text input


# Send prompt to LLM


# Display structured output


# Deliverable:

# Python script or notebook


# Example result


from transformers import pipeline

# Load LLM model
generator = pipeline("text-generation", model="gpt2")

def ai_research_assistant(article):

    prompt = f"""
You are an AI Research Assistant.

Read the article and generate:

1. Short Summary
2. Key Insights (3 points)
3. One Actionable Recommendation

Article:
{article}
"""

    result = generator(prompt, max_length=200)

    print("\n----- AI RESEARCH ASSISTANT OUTPUT -----\n")
    print(result[0]["generated_text"])


# Example article input
article = """
Artificial Intelligence is rapidly transforming industries such as healthcare,
education, and finance. Machine learning algorithms help analyze large datasets
and support decision-making. AI-powered systems assist doctors in diagnosing
diseases and enable personalized learning for students. However, ethical issues
such as bias in algorithms and data privacy must be addressed.
"""

# Run the assistant
ai_research_assistant(article)


```
