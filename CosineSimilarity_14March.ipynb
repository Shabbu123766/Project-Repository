```
# Simple cosine similarity example
import numpy as np

def cosine_similarity(a, b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

a = np.array([0.2,0.8,-0.3,0.5])
b = np.array([0.1,0.9,-0.2,0.4])

print(cosine_similarity(a,b))
```

```
# Scenario: Healthcare Chatbot – Patient Symptom Matching
# Imagine you’re designing a healthcare chatbot that helps patients find possible conditions based on their symptoms. Both patients and conditions are represented as vectors of features (e.g., fever, cough, fatigue, headache).
# - Patient’s symptom vector:
# [0.2,0.8,-0.3,0.5] → mild fever, strong cough, no fatigue, moderate headache.
# - Condition’s symptom profile vector:
# [0.1,0.9,-0.2,0.4] → mild fever, strong cough, little fatigue, moderate headache.
# The chatbot uses cosine similarity to measure how closely the patient’s symptoms align with the condition’s profile.

# Healthcare Chatbot – Patient Symptom Matching using Cosine Similarity

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Patient symptom vector
# [fever, cough, fatigue, headache]
patient_vector = np.array([[0.2, 0.8, -0.3, 0.5]])

# Condition symptom profile vector
condition_vector = np.array([[0.1, 0.9, -0.2, 0.4]])

# Calculate cosine similarity
similarity = cosine_similarity(patient_vector, condition_vector)

print("Patient Symptom Vector:", patient_vector)
print("Condition Symptom Vector:", condition_vector)
print("Cosine Similarity Score:", similarity[0][0])

# Decision example
if similarity[0][0] > 0.8:
    print("Result: High similarity – Condition may match patient's symptoms.")
else:
    print("Result: Low similarity – Condition may not match.")
```

```
# # 🛒 Scenario: E‑Commerce Product Recommendations
# # Imagine you’re building a smart recommendation engine for an online store.
# # - Problem: Customers often search with phrases like “comfortable running shoes”,
#  but the catalog has items
# #  titled “lightweight sneakers for jogging”. Traditional keyword
#  search struggles to connect these.
# # - Solution: Use embeddings to capture semantic meaning of product
# descriptions and customer queries.

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Customer query embedding (semantic meaning of "comfortable running shoes")
query_embedding = np.array([0.1, 0.8, 0.5])

# Product catalog embeddings (semantic meaning of product descriptions)
product_embeddings = {
    "lightweight sneakers for jogging": np.array([0.2, 0.7, 0.6]),
    "formal leather shoes": np.array([0.9, 0.1, 0.2]),
    "sandals for summer": np.array([0.3, 0.4, 0.5])
}

# Compute similarity between query and each product
for product, embedding in product_embeddings.items():
    similarity = np.dot(query_embedding, embedding)  # fast similarity check
    cosine_sim = cosine_similarity([query_embedding], [embedding])[0][0]  # normalized similarity
    print(f"{product} → Dot Product: {similarity:.3f}, Cosine Similarity: {cosine_sim:.3f}")

# Recommendation: pick the product with highest similarity
best_match = max(product_embeddings.items(), key=lambda x: np.dot(query_embedding, x[1]))
print("\nRecommended Product:", best_match[0])
```

```
# Scenario: Online Learning Platform Recommendations
# Imagine you’re building a smart recommendation engine for an online education platform.
# - Problem: Students often search with phrases like “beginner-friendly Python tutorials”, but the course catalog has items titled “Introduction to Programming with Python”. Traditional keyword search struggles to connect these because the words don’t match exactly.
# - Solution: Use embeddings to capture the semantic meaning of both student queries and course descriptions. By comparing embeddings with dot product or cosine similarity, the system can recommend the most relevant courses even when the wording is different.

# Online Learning Platform Recommendation System

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Example embeddings (simulated vectors)
# In real systems these come from models like BERT, OpenAI embeddings, etc.

courses = {
    "Introduction to Programming with Python": [0.9, 0.1, 0.3, 0.8],
    "Advanced Data Science with Python": [0.7, 0.4, 0.6, 0.9],
    "Web Development with JavaScript": [0.2, 0.9, 0.3, 0.2],
    "Machine Learning Basics": [0.8, 0.3, 0.7, 0.6]
}

# Student search query embedding
query = np.array([[0.85, 0.15, 0.35, 0.75]])
# represents: "beginner-friendly Python tutorials"

best_course = None
highest_similarity = -1

for course, embedding in courses.items():

    course_vector = np.array([embedding])

    similarity = cosine_similarity(query, course_vector)[0][0]

    print(f"Similarity with '{course}': {similarity:.3f}")

    if similarity > highest_similarity:
        highest_similarity = similarity
        best_course = course

print("\nRecommended Course:")
print(best_course)
```

```
# Scenario: Hospital FAQ Assistant
# A hospital wants to build a chatbot that helps patients quickly find answers to common questions.
# - Problem: Patients ask questions in different ways. For example:
# - “How do I book an appointment?”
# - “What’s the process to schedule a doctor visit?”
# - “Can I see a physician tomorrow?”
# Traditional keyword search might miss these connections.
# - Solution: Convert FAQs and patient queries into embeddings, then measure similarity to find the closest match.

# 👩‍💻 Code in Context
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# -----------------------------
# Step 1: Pretend these are FAQ embeddings
# (In reality, you'd generate them with a model like MiniLM or mpnet)
# Each vector is a simplified 4-dimensional embedding
faq1 = np.array([0.2, 0.8, -0.3, 0.5])   # "How do I book an appointment?"
faq2 = np.array([0.1, 0.9, -0.2, 0.4])   # "Process to schedule a doctor visit"
faq3 = np.array([-0.7, 0.1, 0.9, -0.5])  # "Visiting hours for patients"

# -----------------------------
# Step 2: Student query embedding
# Again, simplified for demonstration
query = np.array([0.19, 0.82, -0.25, 0.48])  # "Can I see a physician tomorrow?"

# -----------------------------
# Step 3: Cosine similarity
# Measures the angle between vectors (direction matters, not magnitude)
def cosine_sim(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

cos_sim1 = cosine_sim(query, faq1)
cos_sim2 = cosine_sim(query, faq2)
cos_sim3 = cosine_sim(query, faq3)

print("Cosine similarities (higher = more similar):")
print("FAQ1:", cos_sim1)
print("FAQ2:", cos_sim2)
print("FAQ3:", cos_sim3)

# -----------------------------
# Step 4: Euclidean distance
# Measures actual geometric distance (lower = more similar)
dist1 = np.linalg.norm(query - faq1)
dist2 = np.linalg.norm(query - faq2)
dist3 = np.linalg.norm(query - faq3)

print("\nEuclidean distances (lower = more similar):")
print("FAQ1:", dist1)
print("FAQ2:", dist2)
print("FAQ3:", dist3)

# -----------------------------
# Step 5: Dot product (normalized)
# Fastest metric, equivalent to cosine if vectors are normalized
query_norm = query / np.linalg.norm(query)
faq1_norm = faq1 / np.linalg.norm(faq1)
faq2_norm = faq2 / np.linalg.norm(faq2)
faq3_norm = faq3 / np.linalg.norm(faq3)

dot1 = np.dot(query_norm, faq1_norm)
dot2 = np.dot(query_norm, faq2_norm)
dot3 = np.dot(query_norm, faq3_norm)

print("\nDot products (normalized, same as cosine):")
print("FAQ1:", dot1)
print("FAQ2:", dot2)
print("FAQ3:", dot3)
```
